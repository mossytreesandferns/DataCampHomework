# Aggregating Data

```
small_sorted.describe()
```

```
small_sorted['individuals'].mean()
```

```
### Can also use max() and min() with dates
```

```
# Agg method agg()
def pct30(column):
  return column.quantile(.3)
# pass funtion into agg()
small_sorted['individuals'].agg(pct30)
```

```
small_sorted[['individuals','state_pop']].agg(pct30)
```

```
def pct40(column):
  return column.quantile(.4)
# Passing list of functions
small_sorted['individuals'].agg([pct30,pct40])  
```

```
# Cumulative max, min, product, sum
small_sorted['individuals'].cumsum()
```

```
small_sorted['individuals'].sum()
```
