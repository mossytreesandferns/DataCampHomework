### Twinx

```
sixties = data['1960-01-01':'1969-12-01']
fig, ax = plt.subplots()
ax.plot(sixties.index, sixties['LandAverageTemperature'])
ax.plot(sixties['LandAverageTemperatureUncertainty'],color='orange')
ax.set_xlabel('Time')
ax.set_ylabel('Temp (C)/Uncertainty')
plt.show() # Note how scales are very different
```

```
sixties = data['1960-01-01':'1969-12-01']
fig, ax = plt.subplots()
ax.plot(sixties.index, sixties['LandAverageTemperature'])
ax.tick_params('y', colors='blue')
ax.set_xlabel('Time')
ax.set_ylabel('Temp (C)')
ax2 = ax.twinx()
ax2.plot(sixties['LandAverageTemperatureUncertainty'],color='orange')
ax2.tick_params('y', colors='orange')
ax2.set_xlabel('Time')
ax2.set_ylabel('Uncertainty') 
plt.show()
```

```
def plot_timeseries(axes, x,y, color, xlabel, ylabel):
  axes.plot(x,y,color=color)
  axes.set_xlabel(xlabel)
  axes.set_ylabel(ylabel, color=color)
  axes.tick_params('y',colors=color)
```

```
fig, ax = plt.subplots()

plot_timeseries(ax, sixties.index,sixties['LandAverageTemperature'], 'blue','Time','Temp (C)')
ax2 = ax.twinx()
plot_timeseries(ax2,sixties.index,sixties['LandAverageTemperatureUncertainty'],'green','Time','Temp (C)')
plt.show()
```

### Annotations

```
fig, ax = plt.subplots()

plot_timeseries(ax, sixties.index,sixties['LandAverageTemperature'], 'blue','Time','Temp (C)')

ax2 = ax.twinx()
plot_timeseries(ax2,sixties.index,sixties['LandAverageTemperatureUncertainty'],'green','Time','Temp (C)')
ax2.annotate('important point',xy=[pd.Timestamp('1962-07-01'),.25],xytext=[pd.Timestamp('1961-07-01'),.27],arrowprops={})
plt.show()
```

```
bronze = pd.read_csv('bronze_top5.csv', index_col='Country')
silver = pd.read_csv('silver_top5.csv', index_col='Country')
gold = pd.read_csv('gold_top5.csv', index_col='Country')
bronze, silver, gold
```

```
concat = pd.concat([gold, silver, bronze], axis='columns')
concat.columns = ['Gold', 'Silver','Bronze'] 
concat
```

```
fig, ax = plt.subplots()
ax.bar(concat.index, concat['Gold'])
ax.set_xticklabels(concat.index, rotation=45)
ax.set_xlabel('Country')
#plt.xticks(rotation=45)
ax.set_ylabel('Number of Medals')
plt.show()
```

```
fig, ax = plt.subplots()
ax.bar(concat.index, concat['Gold'])
ax.bar(concat.index, concat['Silver'], bottom=concat['Gold'])
ax.set_xticklabels(concat.index, rotation=45)
ax.set_xlabel('Country')
#plt.xticks(rotation=45)
ax.set_ylabel('Number of Medals')
plt.show()
```

```
# add bronze, with labeled legend
fig, ax = plt.subplots()
ax.bar(concat.index, concat['Gold'], label='Gold')
ax.bar(concat.index, concat['Silver'], bottom=concat['Gold'],label='Silver')
ax.bar(concat.index, concat['Bronze'],bottom=concat['Gold']+concat['Silver'], label='Bronze')
ax.set_xticklabels(concat.index, rotation=45)
ax.set_xlabel('Country')
#plt.xticks(rotation=45)
ax.set_ylabel('Number of Medals')
ax.legend()
plt.show()
```

#### Automating Figures from Data

```
athletes = pd.read_csv('athlete_events.csv')
athletes.head()
```

```
athletes.shape
```

```
sports = athletes['Sport'].unique()
sports
```

```
fig, ax = plt.subplots(figsize=(12,8))

for sport in sports:
  sports_df = athletes[athletes['Sport']==sport]
  ax.bar(sport, sports_df['Height'].mean(), yerr=sports_df['Height'].std()) 
ax.set_ylabel('Height')
ax.set_xticklabels(sports, rotation=90)  

plt.show()
```
