### Adding titles and labels

```
#relplot, catplot== facet grids
# scatterplot, countplot === axes subplots

# assign variable for both
```

```
sns.set_context('paper')
sns.set_palette('PRGn')
g = sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
g.fig.suptitle('Bills and Days of Week',y=1.03)

plt.show()
```

```
g=sns.boxplot(x='day',y='total_bill',data=tips)
g.set_title('Totals Per Day',y=1.03)
plt.show()
```

```
g=sns.catplot(x='day',y='total_bill',data=tips,kind='box',col='sex')
g.fig.suptitle('Totals Per Day',y=1.03)
g.set_titles('{col_name}')
plt.show()
```

```
g = sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
g.fig.suptitle('Bills and Days of Week',y=1.03)
g.set(xlabel='Day of Week',ylabel='Total Bill')
plt.show()
```

```
# rotating tiplabels
g = sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
g.fig.suptitle('Bills and Days of Week',y=1.03)
g.set(xlabel='Day of Week',ylabel='Total Bill')
plt.xticks(rotation=45)
plt.show()
```

```
sns.set_palette('Blues')
g = sns.lineplot(x="model_year", y="mpg", 
                 data=mpg,
                 hue="origin")
g.set_title('MPG vs Model Year')
plt.show()
```
